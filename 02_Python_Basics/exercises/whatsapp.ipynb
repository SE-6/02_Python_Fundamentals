Try to build a little Python program with our current knowledge.
It is a (super simplified) age verification for WhatsApp.
Get a username.
Get the user's age.
When the user's age is above 13, they can use WhatsApp, otherwise not.
Print this, than the user is older than 13:
"<username>, you can use WhatsApp: True"
When below
"<username>, you can use WhatsApp: False"

```python
username = input("Enter your username:")
age = input("Enter your age:")

is_old_enough = int(age) > 13

print(f"{username}, you can use WhatsApp: {is_old_enough}")
```
