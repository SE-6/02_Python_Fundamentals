```python
also_good_games = set(["Baldur's Gate", "Monkey Island", "The Witcher", "Dune"])

print(good_old_games)
print(also_good_games)
print("Union", good_old_games.union(also_good_games))
print("Intersection", good_old_games.intersection(also_good_games))
print("Difference", good_old_games.difference(also_good_games))
print("Symmetric Difference", good_old_games.symmetric_difference(also_good_games))
```

```python
set_a = set([1, 2, 3, 4, 5])
set_b = set([4, 5, 6, 7, 8])

print("Union", set_a.union(set_b))
print("Intersection", set_a.intersection(set_b))
print("Difference", set_a.difference(set_b))
print("Symmetric Difference", set_a.symmetric_difference(set_b))
```

```python
colors = set(["green", "orange", "navy", "purple"])
other_colors = set(["black", "blue", "white"])
print(other_colors.issubset(colors))
print(colors.issuperset(other_colors))
print(colors.isdisjoint(other_colors))
```
