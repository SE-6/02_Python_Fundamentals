```python
# Convert the long string into a list. (We did this before, you can look it up) Ignore the header - slicing is your friend.

temp_days = berlin_november.split("\n")[1:]
print(temp_days)
```

```python
# Save the first day (2025-11-17) in a variable on its own, so we can explore it further.
first_day = temp_days[0]
print(first_day)
```

```python
# (first_day) Convert the data into a list of floats. Get rid of the date at the end.
first_day_values = first_day.split(",")[:-1]
# del first_day_values[-1]
print(max(first_day_values))
first_day_floats = [float(val) for val in first_day_values]
print(first_day_floats)
```

```python
# (first_day) Find a way to get the minimum temperature and the maximum temperature. There are in-built functions for that in Python.
print(min(first_day_floats))
print(max(first_day_floats))
```

```python
# (first day) Let's say, the daylight day is always from 7:00 to 17:00. Calculate the average temperature of the daylight day.
first_daylight_day = first_day_floats[7:18]
average = sum(first_daylight_day) / len(first_daylight_day)
print(round(average, 3))
```
