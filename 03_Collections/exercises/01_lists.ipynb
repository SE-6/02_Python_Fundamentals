```python
colors = ["blue", "red", "green", "pink", "orange"]
print(colors)
```

```python
print(colors[0])
print(colors[-1])
```

```python
print(colors[1:4]) # colon
print(colors[:2])
print(colors[2:])
```

```python
middle_colors = colors[1:4]
# print(middle_colors)
middle_colors[0] = "green"
print(middle_colors)
print(colors)

# copy_colors = colors.copy()
copy_colors = colors[:]
print(copy_colors)
```

```python
"pink" in colors
```

```python
colors.append("black")
print(colors)
colors.insert(2, "chartreuse")
print(colors)
```

```python
colors[:2] = ["yellow", "white"]
print(colors)
```

```python
last_element = colors.pop()
print(last_element)
print(colors)
```

```python
more_colors = ["ocker", "purple", "grey", "green", "red"]

print(colors + more_colors)
print(colors.extend(more_colors))
print(colors)
```

```python
numbers = [6, 1, 8, 3, 9, 4, 2, 9]

numbers.sort()
numbers.reverse()
print(numbers)

print(numbers.count(9))
```

```python
# colors.sort()

sorted_colors = sorted(colors)

print(colors)
print(sorted_colors)
```

```python
print(colors.count("green"))
print(colors.index("pink"))
```

```python
# list of colors > 5 chars

new_list = [ item 
            for item in colors 
            if len(item) > 5
            if item.startswith("c")
              ]

print(new_list)
```
