```python
person = {"name": "Guybrush", "age": 20, "height": 1.85}
print(person)
```

```python
print(person["name"])


print(person.get("email", "Email not available"))
```

```python
print(list(person.keys()))
print(person.values())
print(person.items())
```

```python
if "age" in person:
    print("Key 'age' exists")
else:
    print("Key 'age' does not exist")
```

```python
"Guybrush" in person.values()
```

```python
person["city"] = "Bonn"
print(person)
```

```python
person.update({"occupation": "Pirate", "vice": "pillaging", "city": "Monkey Island"})
print(person)
```

```python
age = person.pop("blablabla", "SOME DEFAULT")
print(age)
print(person)
```

```python
person_name = person.setdefault("name", "John Doe")
print(person_name)
print(person)
```

```python
ship = person.setdefault("ship", "Le Ship")
print(ship)
print(person)
```

```python
squares_map = {num: num * num for num in [1, 2, 3, 4, 5, 6, 7] if num % 2 == 0}

print(squares_map)
```
