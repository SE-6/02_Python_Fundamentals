Here is a Tic Tac Toe board. Let's say, you need 3 in a row to win and we ignore columns for now.
You are playing `x` and it's your turn now. Set your final, winning mark.
Now check each row (separately), if there are 3 x in there.
At last, you can reset the board.


```python
tic_tac_toe_board = [
    ["x", " ", "o"],
    ["o", "o", " "],
    ["x", " ", "x"],
]
```

```python
tic_tac_toe_board[2][1] = "x"
print(tic_tac_toe_board)
```

```python
print("Is first row win:", tic_tac_toe_board[0].count("x") == 3)
print("Is second row win:", tic_tac_toe_board[1].count("x") == 3)
print("Is third row win:", tic_tac_toe_board[2].count("x") == 3)
```

```python
tic_tac_toe_board = [
    ["x", " ", "o"],
    ["o", "o", " "],
    ["x", "x", "x"],
]

rows = range(3)

[winning_row_index] = [
    index for index in rows if tic_tac_toe_board[index].count("x") == 3
]

print("winning_row_index:", winning_row_index)
```
