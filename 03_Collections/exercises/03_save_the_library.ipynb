```python
# Strings and lists
#  split the long string and get each data point into a python list (look up the string method `split` and the newline character)
# user_list = library_data.split("\n")
user_list = library_data.splitlines()
print(user_list)
user_list.sort()
print(user_list)
```

```python
# get the first entry from the list and save it to a variable to experiment with it - make sure, there is no excess white space at the start or end of the string
first_user = user_list[0].strip()
print(first_user)
```

```python
# (first user) turn this data into a list (first_user_data) again with `split`
first_user_data = first_user.split(",")
print(first_user_data)
```

```python
# (first user) get the number of borrowed books from the user data and check if it is greater than 0 - make sure you can actually compare your data to an integer
if int(first_user_data[3]) > 0:
    print("User still got some borrowed books")
    print("something else")
else:
    print("User has no books borrowed")
```

```python
# (first user) create a string for an automated, but personalized email which briefly informs the user about the technical issues at hand and about their amount of borrowed books and when they have to return them
# ['123', 'Guybrush Threepwood', 'mighty@pirate.gov', '1', '1990-10-15']

first_user_notification = f"""Dear {first_user_data[1]},
unfortunately, we had a crash. We want to inform you, that you still ave borrowed {first_user_data[3]}, which are due {first_user_data[4]}."""
print(first_user_notification)
```

```python
# List comprehensions
# create a new list based on the full one, containing only users with borrowed books via list comprehension
user_list = library_data.splitlines()
accessible_users = [user.split(",") for user in user_list]
users_with_books = [user for user in accessible_users if int(user[-2]) > 0]
print(users_with_books)
```

```python
# based on that last list, create another one via list comprehension, containing short string informing the user about their borrowed books
# eg: `user_notes = ["Dear Guybrush Threepwood, your 1 book(s) need to be returned by 1990-10-15", "Dear Rand al'Thor, your 15 book(s) need to be returned by 2026-03-23"...]`
notifications = [
    f"Dear {user[1]}, your {user[3]} book(s) need to be returned by {user[-1]}"
    for user in users_with_books
]

print(notifications)
```
