```python
class Blender:
    things_in_container = ""

    def __init__(self):
        pass

    def blend(self):
        if self.things_in_container:
            return "".join(self.things_in_container)
        else:
            return None


my_blender = Blender()


print(my_blender.blend())

print(my_blender)
```

```python
class Present:
    #  Magic Methods - dunder methods
    def __init__(self, name, price, url):
        if not isinstance(price, (float, int)):
            raise ValueError("Price must be a float")
        self.name = name
        self._price = price
        self.url = url
        self.person = ""

    def add_person(self, person_name):
        self.person = person_name

    # Getter
    def get_price(self):
        return f"{self._price:.2f}€"

    # Setter
    def set_price(self, new_price):
        if not isinstance(new_price, float):
            raise ValueError("Price must be a float")
        self._price = new_price

    def wrap(self):
        print(f"Preparing this {self.name} as a present")

    def __str__(self) -> str:
        return f"""
{self.name}: Price: {self.get_price()}
Found on: {self.url}
For {self.person}
"""


marble_track = Present("Marble Track", 42, "marb.le")
marble_track.add_person("Isaiah")

drone = Present("Drone", 81.02, url="drones.com")

# print(marble_track)

marble_track.wrap()
drone.wrap()
# print(marble_track._price)
# marble_track._price = "nonononono"
# print(marble_track._price)
print(marble_track.get_price())
marble_track.set_price(35.32)
print(marble_track.get_price())

print(isinstance(marble_track, Present))

print(marble_track)
print(drone)
```

```python
class BirthdayPresent(Present):
    def __init__(self, name, price, url, date):
        super().__init__(name, price, url)
        self.date = date

    def give(self):
        print(f"Giving {self.name} at {self.date}")

    def wrap(self):
        print(f"Preparing this {self.name} as a present: Reminder date: {self.date}")


book = BirthdayPresent("Qwert", 20, "book7.de", "2013-03-14")

print(isinstance(book, Present))
print(isinstance(book, BirthdayPresent))

print(book.get_price())
book.wrap()
print(book.date)
book.give()
```
