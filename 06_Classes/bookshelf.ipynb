```python
class Book:
    def __init__(self, title, author, release_date, read=False) -> None:
        self.title = title
        self.author = author
        self.release_date = release_date
        self.read = read

    def __str__(self):
        return f"{self.title} - written by {self.author}"
```

```python
class BookCollection:
    """A shelf for books ðŸ“š"""

    def __init__(self, shelf=None):
        self._verify_shelf(shelf)
        self.shelf = shelf or []

    def _verify_shelf(self, shelf):
        if shelf is not None:
            if not isinstance(shelf, list):
                raise ValueError("shelf must be a list")
            for item in shelf:
                self._verify_book(item)

    def _verify_book(self, thing):
        if not isinstance(thing, Book):
            raise ValueError(f"{thing} is not a book")

    def add_book(self, book):
        self._verify_book(book)
        self.shelf.append(book)

    def mark_as_read(self, book_name):
        for book in self.shelf:
            if book.title == book_name:
                book.read = True
                break

    def __str__(self):
        out = []
        for book in self.shelf:
            book_string = (
                f" {"+" if book.read else "-"} {book.title} - written by {book.author}"
            )
            out.append(book_string)
        return "\n".join(out)
```

```python
my_books = BookCollection([ronja, qwert])
print(my_books)
my_books.mark_as_read("Qwert")
print(my_books)
```
