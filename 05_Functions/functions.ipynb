```python
chess_board = [
    ["♜", "♞", "♝", "♛", "♚", "♝", "♞", "♜"],
    ["♟", "♟", "♟", "♟", "♟", "♟", "♟", "♟"],
    [" ", " ", " ", " ", " ", " ", " ", " "],
    [" ", " ", " ", " ", " ", " ", " ", " "],
    [" ", " ", " ", " ", " ", " ", " ", " "],
    [" ", " ", " ", " ", " ", " ", " ", " "],
    ["♙", "♙", "♙", "♙", "♙", "♙", "♙", "♙"],
    ["♖", "♘", "♗", "♕", "♔", "♗", "♘", "♖"],
]


def print_chess_board():
    print("  a b c d e f g h")
    print("  ---------------")
    for index, row in enumerate(chess_board):
        print(f"{8-index}|{" ".join(row)}|{8-index}")
    print("  ---------------")
    print("  a b c d e f g h")


print_chess_board()
chess_board[4][3] = "♙"
chess_board[6][3] = " "
print_chess_board()
```

```python
def add_and_substract(num1, num2):
    a = num1 + num2
    b = num1 - num2
    return a, b


print(add_and_substract(30, 20))

added, substracted = add_and_substract(40, 20)

print(added, substracted)


div, mod = divmod(42, 12)
print(div, mod)
```

```python
def divide(num1, num2):
    try:
        return num1 / num2

    except TypeError:
        print("That wasn't a number")
        return None
    except ZeroDivisionError as zero_err:
        print("0H-N0! I don't know how to divide by zero")
        print(zero_err)
        return None
    except:
        print("I didn't know that this could happen!")
        return None


# print(divide(52, 4))
print(divide(52, 0))
# print(divide(0, 1000000))
# print(divide("GREMLIN", 32))
```
