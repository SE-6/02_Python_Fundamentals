```python
# Let's create a little chess board for the terminal (or the Python Notebook)
# Task 1:
# - create a variable chess_board which is a list of lists
# - each inner list is a row on a chess board, consisting of 8 strings, the square.
# - Each string is either a space " " or a Unicode chess piece, eg. "♟"
# Unicode Chess Pieces:
# White: ♔ ♕ ♖ ♗ ♘ ♙
# Black: ♚ ♛ ♜ ♝ ♞ ♟

chess_board = [
    ["♜", "♞", "♝", "♛", "♚", "♝", "♞", "♜"],
    ["♟", "♟", "♟", "♟", "♟", "♟", "♟", "♟"],
    [" ", " ", " ", " ", " ", " ", " ", " "],
    [" ", " ", " ", " ", " ", " ", " ", " "],
    [" ", " ", " ", " ", " ", " ", " ", " "],
    [" ", " ", " ", " ", " ", " ", " ", " "],
    ["♙", "♙", "♙", "♙", "♙", "♙", "♙", "♙"],
    ["♖", "♘", "♗", "♕", "♔", "♗", "♘", "♖"],
]
# print("♜", "♞", "♝", "♛", "♚", "♝", "♞", "♜")

print(chess_board[6][3])
```

```python
# Task 2:
# - print the board in a formatted way, as in the screenshot below
# - do not add the row and column annotations to your original chess_board
# - use a loop to print the rows

print("  a b c d e f g h")
print("  ---------------")
for index, row in enumerate(chess_board):
    print(f"{8-index}|{" ".join(row)}|{8-index}")
    # print(*row)
    # print("-" * 30)
print("  ---------------")
print("  a b c d e f g h")
```

```python
# Task 3:
# - make an opening move by reassigning a piece in your chess board to another square
# - assign " " to the square where the piece stood before
# - print the board again

# chess_board[4][3] = chess_board[6][3]
chess_board[4][3] = "♙"
chess_board[6][3] = " "

# chess_board[4][3], chess_board[6][3] = chess_board[6][3], chess_board[4][3] # swap


print("  a b c d e f g h")
print("  ---------------")
for index, row in enumerate(chess_board):
    print(f"{8-index}|{" ".join(row)}|{8-index}")
print("  ---------------")
print("  a b c d e f g h")
```
