```python
hour_of_day = 11
weather_is_good = False

if hour_of_day < 10 and weather_is_good:
    print("Good morning")
elif hour_of_day < 14 and weather_is_good:
    print("Good day, indeed")
elif hour_of_day < 14 and not weather_is_good:
    print("Nasty day")
elif hour_of_day < 18:
    print("Good afternoon")
else:
    print("Good evening")


print("END")
# logical and -> &&
# Python: and
```

```python
user_keybord_pressed = input()

match user_keybord_pressed:
    case "w":
        print("Move forward")
    case "s":
        print("Move backwards")
    case "a":
        print("Move left")
    case "d":
        print("Move right")
    case _:
        print("default")
```
